```python
from sklearn.datasets import load_iris
# SciKit Learn - library.
```

```python
iris = load_iris()
# iris is a dictionary-like object.
# load_iris() is a function under sklearn.datasets
print(iris)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(iris.data, columns = iris.feature_names)
# Creates a dataframe with iris.data array contents and gives 
# column names from iris.feature_names array.
```

```python
data.head()
# Prints 1st 5 rows 
```

```python
print(len(data))
```

```python
data['Species'] = pd.DataFrame(iris.target)
# Store target values in column species.
```

```python
data.head()
```

```python
x = data.iloc[:,:-1] 
# All columns except the last column will be stored in x
y = data.iloc[:,-1]
# Only the last column will be stored in y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
# Split arrays or matrices into random train and test subsets
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = "entropy", splitter = "best")

# Criterion :
# It is used to evaluate the feature importance.
# Default is gini, but we can also use entropy. Based on this
# the model will define the importance of each feature for the classification.

# Splitter:
# The splitter is used to decide which feature and which threshold is used.
# Using best, the model if taking the feature with the highest importance.
# Using random, the model if taking the feature randomly but with the same distribution
# (in gini, proLine have an importance of 38% so it will be taken in 38% of cases)
```

```python
model_fit = model.fit(x_train, y_train)
y_pred = model.predict(x_test)

# Model fitting is a measure of how well a machine learning model 
# generalizes to similar data to which it was trained
```

```python
from sklearn.metrics import accuracy_score
# In multi-label classification, this function computes accuracy 
# of subset 
print(accuracy_score(y_pred, y_test).round(2) * 100)
# Calculate accuracy, round it by 2 and print percentage
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
# This command displays the plot within the notebook itself
```

```python
from sklearn.tree import plot_tree
# Decision tree is a non-parametric supervised learning method 
# used for classification & regression. The goal is to create a model that predicts the 
# value of a target variable inferred from the data features
```

```python
# Plot decision tree
plt.figure(figsize=(25, 30))
plot_tree(model_fit, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
